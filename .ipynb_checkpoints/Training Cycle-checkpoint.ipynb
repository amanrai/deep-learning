```python
%matplotlib inline
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
import json
_data = None
sns.set_style("whitegrid")
with open("./training_cycle.json", "r") as f:
    _data = json.loads(f.read())
trg = _data["training_losses"]
data = _data["validation_losses"]
acc = _data["validation_accuracy"]
evid = np.round(_data["evidence_accuracy"],2)
labels = len(data)
width = len(data)
if (width < 12):
    width = 12
plt.figure(figsize=(width, 8))
ticks = range(width)
while (len(data) < width):
    data.append(None)
    trg.append(None)
    acc.append(None)
import pandas as pd
df = pd.DataFrame({
    "validation":data,
    "training":trg,
    "validation accuracy":acc
})
g = sns.lineplot(data=df, marker="o")
g.set(xticks = range(len(data)))
g.set(xlabel = "epochs")
axes = g.axes
axes.set_ylim(0, 1.25)
g.set(ylabel = "losses (categorical crossentropy)")
for i in range(labels):
    g.text(i+0.2, data[i], 
           "evid:" + str(evid[i]), 
           horizontalalignment='left', 
           size='medium', 
           color='gray', weight='light')
    
```
